```python
import pandas as pd
import matplotlib.pyplot as plt

# Upload file manually
from google.colab import files
uploaded = files.upload()

# Load file
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

print(df.head())

# Basic exploration
print(df.head())
print(df.info())
print(df.describe())

# Clean data
df = df.dropna()

# Revenue by Country
country_sales = df.groupby("COUNTRY")["SALES"].sum().sort_values()

# Plot
country_sales.plot(kind="barh")
plt.title("Sales by Country")
plt.xlabel("Revenue")
plt.show()
```

**TOP Product Analysis**

```python
top_products = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values()

top_products.plot(kind="bar")
plt.title("Top Product Sales")
plt.show()
```

**Monthly Trend**

```python
monthly_sales = df.groupby("MONTH_ID")["SALES"].sum()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.show()
```

Insights:
- Australia shows strong revenue concentration
- Classic Cars generate highest sales
- Sales peak observed in November
